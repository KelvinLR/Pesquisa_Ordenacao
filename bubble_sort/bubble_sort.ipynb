## Objetivo da Atividade

Analisar o desempenho do algoritmo de ordenação Bubble Sort medindo o tempo de execução para diferentes tamanhos de entrada e entendendo o impacto do tamanho dos dados no tempo de execução.

```python
# Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
# Ordenação utilizando bubble sort
def bubble_sort(list):
    for i in range(len(list)):
      for j in range(len(list)-1):
        if(list[j] > list[i]):
          aux = list[i]
          list[i] = list[j]
          list[j] = aux

# Função para retornar o tempo de ordenação com o bubble sort
def get_sort_time(list):
  start = time.perf_counter()
  bubble_sort(list)
  end = time.perf_counter()
  total_time = end - start
  return total_time
```

```python
# Definindo o tamanho dos vetores
sizes = [10000, 20000, 30000, 40000, 50000]

# Inicializando o vetor 'arrays' vazio
arrays = []

for size in sizes:
  # Gerando vetor aleatório com as posições definidas
  my_list = np.random.randint(0, 10*size,size=size)
  # Adicionando ao vetor 'arrays'
  arrays.append(my_list)
```

```python
times_bubblesort_general_case = []

# Ordenando os vetores gerados (caso geral)
for array in arrays:
  # Obtendo tempo de execução
  times_bubblesort_general_case.append(get_sort_time(array))
```

```python
times_bubblesort_best_case = []

# Obtendo tempo de execução para o melhor caso
for array in arrays:
  times_bubblesort_best_case.append(get_sort_time(array))
```

```python
times_bubblesort_worst_case = []

# Ordenando os vetores gerados (pior caso)
for array in arrays:
    # Invertendo vetor
    reversed_array = np.flip(array)
    # Obtendo tempo de execução
    times_bubblesort_worst_case.append(get_sort_time(reversed_array))
```

```python
plt.style.use('seaborn-v0_8')

# Plotando gráfico com o tempo de execução do caso geral, melhor caso e pior caso
plt.plot(sizes, times_bubblesort_best_case, label='Melhor caso')
plt.plot(sizes, times_bubblesort_general_case, label='Caso geral')
plt.plot(sizes, times_bubblesort_worst_case, label='Pior caso')
plt.legend(loc='upper left')
plt.xlabel('Tamanho')
plt.ylabel('Tempo de Execução')
plt.title('Tempo de Execução vs. Tamanho do Vetor (Bubble Sort)')
plt.show()
```
